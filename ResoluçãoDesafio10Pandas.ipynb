# 8. Quem tem maior pressão sanguínea média, homens ou mulheres? (Considere a média)

```python
import pandas as pd

# Carregar o dataset
df = pd.read_csv('1KBtJf20QHGyrNOTEGeq_saude_do_sono_estilo_vida.csv')
```

```python
df.rename(columns={'ID': 'Identificador','Pressão sanguíneaaaa': 'Pressão sanguínea', 'Ocupação': 'Profissão','Categoria BMI': 'Categoria IMC'}, inplace=True)
# Exibir algumas linhas para conferir os dados
df.head()
```

```python
# Função para calcular a média da pressão sanguínea
def calcular_media_pressao(df, genero):
    # Filtrar por gênero
    filtro = df[df['Gênero'] == genero]

    # Separar a pressão sanguínea em sistólica e diastólica
    pressao = filtro['Pressão sanguínea'].str.split('/', expand=True)
    pressao_sistolica = pressao[0].astype(float)
    pressao_diastolica = pressao[1].astype(float)

    # Calcular a média
    media_sistolica = pressao_sistolica.mean()
    media_diastolica = pressao_diastolica.mean()

    return media_sistolica, media_diastolica
```

```python
# Calcular a média para homens e mulheres
media_homens_sistolica, media_homens_diastolica = calcular_media_pressao(df, 'Homem')
media_mulheres_sistolica, media_mulheres_diastolica = calcular_media_pressao(df, 'Mulher')
```

```python
# Exibir os resultados
print(f"Média da pressão sanguínea para homens: Sistólica = {media_homens_sistolica:.2f}, Diastólica = {media_homens_diastolica:.2f}")
print(f"Média da pressão sanguínea para mulheres: Sistólica = {media_mulheres_sistolica:.2f}, Diastólica = {media_mulheres_diastolica:.2f}")
```

```python
# Determinar quem tem a maior média
if media_homens_sistolica > media_mulheres_sistolica:
    print("Homens têm maior pressão sanguínea média (sistólica).")
elif media_homens_sistolica < media_mulheres_sistolica:
    print("Mulheres têm maior pressão sanguínea média (sistólica).")
else:
    print("Homens e mulheres têm a mesma pressão sanguínea média (sistólica).")

if media_homens_diastolica > media_mulheres_diastolica:
    print("Homens têm maior pressão sanguínea média (diastólica).")
elif media_homens_diastolica < media_mulheres_diastolica:
    print("Mulheres têm maior pressão sanguínea média (diastólica).")
else:
    print("Homens e mulheres têm a mesma pressão sanguínea média (diastólica).")
```

# 10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
# Separar os dados em dois grupos:
# 1. Frequência cardíaca > 70
# 2. Frequência cardíaca <= 70
grupo_acima_70 = df[df['Frequência cardíaca'] > 70]
grupo_ate_70 = df[df['Frequência cardíaca'] <= 70]
```

```python
print(grupo_acima_70)
print(grupo_ate_70 )
```

```python
# Calcular a média de passos diários para cada grupo
media_passos_acima_70 = grupo_acima_70['Passos diários'].mean()
media_passos_ate_70 = grupo_ate_70['Passos diários'].mean()
```

```python
# Exibir os resultados
print(f"Média de passos diários para pessoas com frequência cardíaca > 70: {media_passos_acima_70:.2f}")
print(f"Média de passos diários para pessoas com frequência cardíaca <= 70: {media_passos_ate_70:.2f}")
```

```python
# Comparar as médias
if media_passos_acima_70 > media_passos_ate_70:
    print("Pessoas com frequência cardíaca acima de 70 dão mais passos por dia.")
elif media_passos_acima_70 < media_passos_ate_70:
    print("Pessoas com frequência cardíaca menor ou igual a 70 dão mais passos por dia.")
else:
    print("Ambos os grupos dão a mesma quantidade média de passos por dia.")
```
